```python
category_mapping = {'Fiction' : "Fiction",
 'Juvenile Fiction': "Children's Fiction",
 'Biography & Autobiography': "Nonfiction",
 'History': "Nonfiction",
 'Literary Criticism': "Nonfiction",
 'Philosophy': "Nonfiction",
 'Religion': "Nonfiction",
 'Comics & Graphic Novels': "Fiction",
 'Drama': "Fiction",
 'Juvenile Nonfiction': "Children's Nonfiction",
 'Science': "Nonfiction",
 'Poetry': "Fiction"}
```

```python
books["simple_categories"]=books["categories"].map(category_mapping)
```

```python
books[~books["simple_categories"].isna()]
```

```python
from transformers import pipeline
cat = ["Fiction","Nonfiction"]
pipe = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
seq = books.loc[books["simple_categories"]== "Fiction","description"].reset_index(drop=True)[7]
```

```python
pipe(seq,cat)
```

```python
import numpy as np
max_index = np.argmax(pipe(seq,cat)["scores"])
max_label = pipe(seq,cat)["labels"][max_index]
max_label
```

```python
def pred(seq,cat):
    prediction = pipe(seq,cat)
    max_index = np.argmax(prediction["scores"])
    max_label = prediction["labels"][max_index]
    return max_label

```

```python
from tqdm import tqdm
actual_cat= []
predicted_cat = []
for i in tqdm(range(0,400)):
    s = books.loc[books["simple_categories"]=="Fiction","description"].reset_index(drop=True)[i]
    p = pred(s,cat)
    predicted_cat += [p]
    actual_cat +=["Fiction"]
```

```python
for i in tqdm(range(0,400)):
    s = books.loc[books["simple_categories"]=="Nonfiction","description"].reset_index(drop=True)[i]
    p = pred(s,cat)
    predicted_cat += [p]
    actual_cat +=["Nonfiction"]
```

```python
pd_df = pd.DataFrame({"actual":actual_cat,"predicted":predicted_cat})
```

```python
pd_df
```

```python
pd_df["correct"]= np.where(pd_df["actual"]==pd_df["predicted"],1,0)
```

```python
pred_lt = []
isb = []
missing_cat = books.loc[books["simple_categories"].isna(),["isbn13","description"]].reset_index(drop=True)
```

```python
for i in range(len(missing_cat)):
    d = missing_cat["description"][i]
    pred_lt += [pred(d,cat)]
    isb += [missing_cat["isbn13"][i]]
```

```python
missing_predicted_df = pd.DataFrame({"isbn13": isb, "predicted_categories": pred_lt})
```

```python
missing_predicted_df
```

```python
books = pd.merge(books, missing_predicted_df, on="isbn13", how="left")
books["simple_categories"] = np.where(books["simple_categories"].isna(), books["predicted_categories"], books["simple_categories"])
books = books.drop(columns = ["predicted_categories"])
```
