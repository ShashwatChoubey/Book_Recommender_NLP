```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shashwatchoubey/books-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/final_books_data.csv")
```

```python
books
```

```python
books.dropna(subset=["isbn13","title"],inplace=True)
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(),cbar=False,ax=ax)

plt.xlabel("Missing Values")
plt.ylabel("Columns")
plt.show()
```

```python
import numpy as np
books["missing_description"] = np.where(books['description'].isna(),1,0)
books["age"] = 2024 - books["published_year"]
```

```python
columns_of_interest = ["num_pages","age","missing_description","average_rating"]
correlation_matrix = books[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(10,10))
heatmap = sns.heatmap(correlation_matrix,annot=True,fmt=".2f",cmap="coolwarm",cbar_kws={"label":"Spearman correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()
```
